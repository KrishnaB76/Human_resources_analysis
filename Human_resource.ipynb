```python
#creating a column - age of employees
df['age']=2024-pd.DatetimeIndex(df['birthdate']).year.astype(int)
df.head(5)
```

```python
#What is the gender breakdown of employees in the company?
df.groupby(by='gender').agg(['count'])['Id']
```

```python
#What is the race/ethnicity breakdown of employees in the company?
df.groupby(by='race').count()['Id']
```

```python
#What is the age distribution of employees in the company?
#creation of age group
for i in range(df["Id"].count()):
    a = df.iloc[i]['age']
    if a >= 18 and a < 25:
        df.at[i, 'age_group'] = "18-24"
    elif a >= 25 and a < 35:
        df.at[i, 'age_group'] = "25-34"
    elif a >= 35 and a < 45:
        df.at[i, 'age_group'] = "35-44"
    elif a >= 45 and a < 55:
        df.at[i, 'age_group'] = "45-54"
    elif a >= 55 and a < 65:
        df.at[i, 'age_group'] = "55-64"
    else:
        df.at[i, 'age_group'] = "65+"
#grouping by age group
df.groupby(by="age_group").count()['Id']
```

```python
#How many employees work at headquarters versus remote locations?
df.groupby(by="location").count()['Id']
```

```python
#What is the average length of employment for employees who have been terminated?
a=0.0
b=0.0
for i in range(df['Id'].count()-1):
  if(type(df.iloc[i]['term_date'])==str and int(df.iloc[i]['term_date'][0:4]) <=2024):
    a+= (2024-int(df.iloc[i]['term_date'][0:4]))
    b+=1
print(round((a/b),2))


```

```python
#How does the gender distribution vary across departments?
print(df.groupby(by=['department','gender']).count()['Id'])
```

```python
#What is the distribution of job titles across the company?
df.groupby(by="job_title").count()['Id']
```

```python
#What is the distribution of employees across locations by state?
df.groupby(by='location_state').count()['Id']
```

```python
#How has the company's employee count changed over time based on hire and term dates?
tc=df.groupby(by=pd.DatetimeIndex(df['hire_date']).year).count()[['Id','term_date']]
tc= tc.rename(columns={'Id':'number_of_hires','term_date':'number_of_terminations'})
for i in range(tc['Id'].count()-1):
 tc['net_change']=tc['Id']-tc['term_date']
 tc['net_change%']= ((tc['Id']-tc['term_date'])/tc['Id'])*100
tc
```

```python
#What is the tenure distribution for each department?
for i in range(df['Id'].count()-1):
  a=0
  a=int(str(df.iloc[i]['hire_date'])[0:4])#extracting hire year
  if(df.iloc[i]['term_date']== str):
    df['term_year']=str(df.iloc[i]['term_date'])[0:4]#extracting termination year
    df['tenure']=df['term_year']-a
df.groupby(by='department').mean()['tenure']


```
